## Infraestrutura

As salas com terminais da Faculdade de Ciências e Tecnologia (FCT) onde decorrem os laboratórios de Introdução à programação científica, estão equipadas com computadores ligados à rede UAlg e acesso a um conjunto de recursos. Esta organização é muito prática, pois assim teremos os programas necessários para trabalhar.

Desta forma, é possível oferecer aos alunos a capacidade de trabalhar desde o início com as aplicações convenientes, não sendo necessário que os alunos usem os seus computadores pessoais e os tenham que trazer para as aulas práticas. Ainda assim, muitos alunos preferem usar também o seu computador pessoal, porque pode ser mais conveniente ter sempre acesso ao sistema e ambiente de trabalho. Para partilhar ficheiros entre o sistema pessoal e o usado nas aulas práticas sugere-se um sistema de partilha de ficheiros como o OneDrive (onde os alunos têm 1 TB).

Em todo o caso, todos os alunos devem familiarizar-se com o sistema da Universidade do Algarve, que é a referência a ter em conta.

## Identificar as ferramentas a utilizar


Na área de trabalho acessível pelos computadores do laboratório já temos todo o software de que precisamos. 

Talvez queira instalar no seu computador a plataforma [Anaconda][2] ou pelo menos o [Python][1] e o [Spyder][3]. Visite as páginas indicadas, para se situar e depois instale nos seus computadores.

[1]: https://www.python.org/downloads/ "Python"
[2]: https://www.continuum.io/downloads "Anaconda"
[3]: https://www.spyder-ide.org/ "Spyder"

Alguns alunos poderão preferir usar outra plataforma em vez de Anaconda. Não é proibido, claro, mas haverá mais uma ferramenta a aprender e, tal como a escolha é individual, a aprendizagem também será. 


## Criar uma área de trabalho para Introdução à programação científica

Na pasta *Documents* da sua área crie uma pasta IPC_1920. Dentro desta, crie uma pasta geral.

Guarde tudo o que tiver a ver com a nossa cadeira nessa pasta!

Abra o IDE Spyder e crie um pequeno texto, como quiser e com o que quiser. Guarde-o na sua pasta geral.

E, se for essa a sua opção, faça isto no seu computador pessoal também. Mantenha os dois computadores “iguais”, por assim dizer.

Se estiver a usar o computador da sala, saia da sessão. Depois volte a entrar, noutro computador. Verifique que o seu ficheiro está lá.


## Primeiros passos

No menu *Iniciar* do seu computador, localize o *Jupyter QTConsole*
 
A linha de comando do python é representada por '>>>', designada *prompt*. Na linha de comando, a introdução de uma instrução completa é seguida pela sua execução.

Usando o exemplo mais famoso da informática, escreva:

    print ("Olá mundo")
e pressione *Enter*. Fará com que seja apresentado:

    Olá mundo

Uma das utilizações típicas da linha de comando do python é como calculadora. Teste os exemplos apresentados na aula teórica.

## Notebook

Da página da Tutoria descarregue o ficheiro *T1.ipynb* para a sua área.
A partir do menu *Iniciar* abra agora o *Jupyter Notebook*. No separador do explorador que foi aberto, localize o ficheiro descarregado e abra-o.

Vá pressionando as teclas *Shift*+*Enter* para executar as células. Experimente mudar o seu conteúdo.
